slimming down the eras to one in 10.
Syntax: 
- train['era'].unique() grabs the eras column and filters for unique entries. We then filter for every 10th entry
- then we downsample to one in 10 (train [...] can take in a boolean mask)

```python
print(len(train), len(features))

train = train[train["era"].isin(train["era"].unique()[::1000])]
print(len(train))
```

- feature_list + ['target'] concatenates two list (note that target is not a feature)
- train[feature_list + ['target']] grabs only the columns in the concatenated list
- .coor(method="spearman") measures the spearman correlation from every column to every other column
- we only grab the correlation with the target
- note that the column also includes the correlation of target itself, so we drop it. 
- then we sort the correlation values in descending order, and grab the head and tail. 

```python
# Which features correlate with target at all?
feature_corrs = train[features + ["target"]].corr(method="spearman")["target"]
feature_corrs = feature_corrs.drop("target").sort_values(ascending=False)
print(feature_corrs.head(20))   # most predictive
print(feature_corrs.tail(20))   # negatively predictive
```

Feature_era_coor: 
- groups the training data by era
- For each era, calculate the correlation of a given column (feature) with the target column
- output: dataframe with rows = eras, columns = features, cellval = correlation

.mean() calculates the average correlation, then we sort and get the features with the most corelation
.std does the same, but we get the most stable features (smallest std)


```python
# How stable is each feature's correlation across eras?
def feature_era_corr(train, features):
    return train.groupby("era").apply(
        lambda x: x[features].corrwith(x["target"], method="spearman")
    )

era_feature_corrs = feature_era_corr(train, features)
```

```python
print(era_feature_corrs.mean().sort_values(ascending=False).head(20))
```

```python
print(era_feature_corrs.std().sort_values().head(20))  # most stable features
```

Missing features are given a 2, so their variance is NaN

```python
# Check how many eras each feature actually has data in
# (i.e., is not all the same value)
def feature_coverage(train, features):
    def has_variance(x):
        return (x[features].nunique() > 1).astype(int) #returns a series of 0 and 1, one number per feature
    # for each era, calculate if a given feature has variance.
    # then we take the average over all eras
    return train.groupby("era").apply(has_variance).mean() 

feature_sharpe = (
    era_feature_corrs.mean() / era_feature_corrs.std()
).dropna().sort_values(ascending=False) # calculate the sharpe ratio
coverage = feature_coverage(train, feature_sharpe)
print(coverage.sort_values().head(20))  # features with lowest coverage
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from config import Config
from data import load_dataset

config = Config(feature_set="medium")
train, features = load_dataset(config, "train")
train = train[train["era"].isin(train["era"].unique()[::1000])]

feature_corrs = train[features + ["target"]].corr(method="spearman")["target"]
feature_corrs = feature_corrs.drop("target").sort_values(ascending=False)

top_features = feature_corrs.abs().sort_values(ascending=False).head(50).index.tolist()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train[top_features])
print('train rows:', X_train_scaled.shape[0])
```

```python
ridge = Ridge(alpha=100)
ridge.fit(X_train_scaled, train['target'])
last_train_era = int(train['era'].unique()[-1])

validate(ridge, top_features, last_train_era)
```
